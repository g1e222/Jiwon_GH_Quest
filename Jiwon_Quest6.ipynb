```python
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer 
lbc=load_breast_cancer()


print(lbc.keys())
```

```python
X = lbc['data'][:,0].reshape(-1,1)
y = lbc['target']

X, y, lbc.DESCR
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

X = X[:,0].reshape(-1,1)
y = (y==0).astype(np.int)

log_reg = LogisticRegression()
log_reg.fit(X,y)
#꽃받침 길이를 기준으로 Sentosa/Not Sentosa를 분류하고 싶다
#꽃받침 길이가 2이면 Not Sentosa, 꽃받침 길이가 4.5이면 Sentosa
```

```python
X_new = np.linspace(6,30,1000).reshape(-1,1)
y_prob = log_reg.predict_proba(X_new)
plt.plot(X_new, y_prob[:,1], 'g-', label='positive')
plt.plot(X_new, y_prob[:,0], 'b-', label='negative')
```
