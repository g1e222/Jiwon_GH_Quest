```python
from nltk.corpus import gutenberg
files_en = gutenberg.fileids()      # Get file ids
files_en
```

```python
doc_en = gutenberg.open('shakespeare-hamlet.txt').read()
doc_en
```

```python
# nltk - 1. tokenize
tokens_en = nltk.word_tokenize(doc_en) #token단위로 쪼개고 리스트로 객체화
print(tokens_en)
```

```python
no_n_doc = re.sub("\n", ' ', doc_en)
sentences=nltk.sent_tokenize(no_n_doc)
len(sentences)
sentences[0:5]
```

```python
en = nltk.Text(tokens_en)
type(en)
```

```python
print(len(en.tokens))
print(len(tokens_en))

print(len(set(en.tokens)))
print(len(set(tokens_en)))
```

```python
# nltk -2. pos_tag
tags_en = nltk.pos_tag(tokens_en)
```

```python
tags_en[:20]
```

```python
#https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html
```

```python
n=[word for word,pos in tags_en if pos in ['NN','NNS','NNP','NNPS']]
#NN Noun, singular or mass
#NNS Noun, plural
#NNP Proper noun, singular
#NNPS Proper noun, plural
```

```python
v=[word for word,pos in tags_en if pos in ['VB','VBD','VBG','VBN''VBP','VBZ']]
#VB Verb, base form
#VBD Verb, past tense
#VBG Verb, gerund or present participle
#VBN Verb, past participle
#VBP Verb, non-3rd person singular present
#VBZ Verb, 3rd person singular present
```

```python
adj=[word for word,pos in tags_en if pos in ['JJ','JJR','JJS']]
#JJ Adjective
#JJR Adjective, comparative
#JJS Adjective, superlative
```

```python
nplot = nltk.FreqDist(n) 
nplot.plot(20, cumulative=False)
```

```python
vplot = nltk.FreqDist(v) 
vplot.plot(20, cumulative=False)
```

```python
adjplot = nltk.FreqDist(adj) 
adjplot.plot(20, cumulative=False)
```

```python
nplot= nltk.FreqDist(n) 
nplot.plot(50, cumulative=False)
```
