### Pandas를 이용한 데이터 불러오기 & 정리

```python
import pandas as pd
import numpy as np
```

```python
CCTV_Seoul = pd.read_csv('CCTV_in_Seoul.csv', encoding = 'utf-8') #인코딩이 안되면 engine = 'python'으로 해보세요!
CCTV_Seoul.head(3)
```

```python
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0] :'지역구별'}, inplace = True)
CCTV_Seoul.head()
```

```python
pop_Seoul = pd.read_excel('population_in_Seoul.xls', 
                          header = 2, parse_cols = 'B, D, G, J, N', encoding = 'utf-8')
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns={pop_Seoul.columns[0]:'지역구별',
                          pop_Seoul.columns[1]:'인구수',
                          pop_Seoul.columns[2]:'한국인',
                          pop_Seoul.columns[3]:'외국인',
                          pop_Seoul.columns[4]:'고령자'},inplace=True)
pop_Seoul.head()
#방금 생성한 pop_Seoul 객체의 칼럼 이름을 위와 같이 만들어 보세요!
```

```python
pop_Seoul.drop([0], inplace=True)
pop_Seoul.head()
#맨 위의 ‘합계’ 행을 삭제해 보세요! (drop 이용)
```

```python
CCTV_Seoul.sort_values(by = '소계', ascending = True).head(5)
```

```python
CCTV_Seoul.sort_values(by = '소계', ascending = False).head(5)
```

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2016년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2014년']) / CCTV_Seoul['2013년도 이전'] * 100
CCTV_Seoul.head(5)
```

```python
CCTV_Seoul.sort_values(by='최근증가율',ascending=False).head(5)
#'최근증가율' 칼럼을 기준으로 내림차순 정렬!
```

```python
data_result=pd.merge(CCTV_Seoul,pop_Seoul,on='지역구별')
data_result.head()
#CCTV_Seoul 객체와 pop_Seoul 객체를 다음과 같이 통합해 보세요! (merge 함수 이용, 만들어진 객체의 이름은 'data_result'으로)
```

```python
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
data_result.head()



#'2013년도 이전', '2014년', '2015년', '2016년' 칼럼을 삭제해 주세요
```

```python
data_result.set_index('지역구별', inplace=True)
data_result.head()
```

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
fp1
#array([기울기, 절편])
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50) #s: 점 사이즈
plt.plot(fx, f1(fx), ls='dashed', lw=3, color ='g')
plt.xlabel('인구수')
plt.ylabel('소계')
plt.grid()
plt.show()
```

```python
data_result['외국인비율']=data_result['외국인']/data_result['인구수']*100
data_result['고령자비율']=data_result['고령자']/data_result['인구수']*100
data_result['오차']=np.abs(data_result['소계']-(fp1[0]*data_result['인구수']+fp1[1]))
data_result.sort_values(by='오차',ascending=False).head(500)

    
#'외국인비율’ = ‘외국인’ / ‘인구수’ * 100, ‘고령자비율’ = ‘고령자’ / ‘인구수’ * 100
#‘오차’ – l소계 – 회귀직선(인구수)l , 오차는 np.abs를 활용
```

```python
#2
#(1)에서 정리한 데이터를 ‘오차’ 컬럼을 기준으로 색깔을 입혀서 시각화 해보고
#이를 통해 무엇을 파악할 수 있는지 주석으로 간략하게 1-2줄로 서술해보기. (plt.colormap, plt.text 로 컬러맵과 이름 추가)
```

```python
import matplotlib.pyplot as plt #matplotlib가 한글폰트를 지원 안해서 미리 처리해주는 과정
import platform                 #아마 요즘 matplotlib는 이런거 안해도 잘 될수도 있어요!

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus']= False

if platform.system() == 'Darwin': #맥os 사용자
    rc('font', family = 'AppleGothic')
elif platform.system() == 'Windows':#윈도우 사용자
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name) #김신=도깨비
```

```python
re_data_result=data_result.sort_values(by='오차', ascending=False)

x=re_data_result['인구수']
y=re_data_result['소계']
name=re_data_result.index #indext: 행

colorbar = data_result['오차']

plt.figure(figsize=(10,7))
plt.scatter(data_result['인구수'], data_result['소계'], c = colorbar, s=50) #s: 점 사이즈
plt.plot(fx, f1(fx), ls='dashed', lw=3, color ='g')
plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.grid()
plt.colorbar()


for i in range (9):
    plt.text(x[i], y[i], name[i], fontsize = 10) #plt.text(x축좌표, y축좌표, 메세지, 폰트사이즈)
    
plt.show()

# 그래프 위의 회귀선에서 멀수록 오차가 큰 데이터임을 알 수 있다. 라벨이 달린 지역구는 회귀선으로 설명되기 어렵다.
    
```
